Problem2:



```python
x = np.arange(-3,4)

# since sigma(Pi) = 1
# sum(xi**2)/a = 1

a = np.sum(x**2)
print("a = ", a)

## Now that we have a, we can procedd to next parts.
p = [(p**2)/a for p in x]
E = np.sum(x*p)
print("Expectation, E = ", round(E,2))

# For cal. Var we would use V(x) = E(x**2) - [E(x)]**2

Var = np.sum(np.square(x)*p) - np.square(E)
print("Variance, V = ", Var)
```
