```python
import pandas as pd
```

Notes:

1. df.isna() will return True/False for each element of df.
2. df[df.isna()] will retrieve only those elements of df which are pd.NA, as , otherwise will retrieve Nan. You can try this using:

3. df.isnull()) will return True/False for each element and any(axis=1) on the returned dataframe will return True/False for each row.

4. df[df.isnull().any(axis=1)] will retrieve the rows having any Nan element. If we use isna() inplace of isnull(), it will give us the same results.

```python
df= pd.DataFrame({'Name':["Jim","Carrie","Chris","Morris"],
        'Gender':["M","F","M",pd.NA],
        "Profession":["Athelete","Tech","Cricketer","Actor"]})
df
```

```python
df.isna()
```

```python
print(df[df.isna()])
```

```python
df[df.isnull().any(axis=1)] 
# here with axis = 1 means it will collapse all columns
```

```python
df.isnull().all(axis=1)
```

# Pandas groupby function

```python
import pandas as pd
student_data = pd.DataFrame({
'school_code': ['s001','s002','s003','s001','s002','s004','s001','s003'],
'class': ['V', 'V', 'VI', 'VI', 'V', 'VI','IX','X'],
'name': ['Alberto Franco','Gino Mcneill','Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill', 'David Parkes','Jane','Martin'],
'age': [12, 12, 13, 13, 14, 12, 15, 20],
'height': [173, 192, 186, 167, 151, 159, 140, 160],
'weight': [35, 32, 33, 30, 31, 32, 40, 50]},
 index=['S1', 'S2', 'S3', 'S4', 'S5', 'S6','S7','S8']) 
student_data
```

```python
#Block a
grouped_data = student_data.groupby(['school_code', 'class'])      
print((grouped_data.size())) 

#Code in block a, prints the number of observations in all the unique 'school_code' and 'class' combinations.
```

```python
#Block b
n = 1                                                              
rows_to_drop=student_data.groupby(['school_code']).tail(n).index
result1 = student_data.drop(rows_to_drop, axis=0)
print(result1)

# Code in block b, prints all the observations of each 'school_code' 
# except the last one observation in each unique school_code.
```
